```python
import pandas as pd
```

```python
df = pd.read_csv('../data/EPA_SmartLocationDatabase_V3_Jan_2021_Final.csv')
df.head()
```

```python
df['STATEFP'] = df['STATEFP'].astype(str).str.zfill(2)
df['COUNTYFP'] = df['COUNTYFP'].astype(str).str.zfill(3)
df['TRACTCE'] = df['TRACTCE'].astype(str).str.zfill(6)
df['BLKGRPCE'] = df['BLKGRPCE'].astype(str)
df.head()
```

```python
# Merge selected columns into one as string dtype
cols = ['STATEFP', 'COUNTYFP', 'TRACTCE', 'BLKGRPCE']
df['GEOID'] = df[cols].apply(lambda row: ''.join(row.values.astype(str)), axis=1)
```

```python
# Move GEOID column
df.insert(3, 'GEOID', df.pop('GEOID'))
```

```python
# Get index number of "NatWalkInd" column
index = df.columns.get_loc('NatWalkInd')
print(index)
```

```python
# Move NatWalkInd column
df.insert(111, 'NatWalkInd', df.pop('NatWalkInd'))
df.head()
```

```python
df = df.drop(['OBJECTID', 'GEOID10', 'GEOID20', 'CSA', 'CSA_Name', 'CBSA', 'Shape_Length', 'Shape_Area'], axis=1)
df.head()
```

```python
df = df[['GEOID', 'STATEFP', 'COUNTYFP', 'TRACTCE', 'BLKGRPCE', 'CBSA_Name', 'CBSA_POP', 'Ac_Total', 'TotPop', 'R_LowWageWk',
         'R_MedWageWk', 'R_HiWageWk', 'R_PCTLOWWAGE', 'NatWalkInd']]
df.head()
```

```python
df["Walkable"] = df["NatWalkInd"].apply(lambda x: "Yes" if x >= 10.51 else "No")
print(df.shape)
df.head()
```

```python
df.to_csv('../data/walkable_tableau.csv')
```
